```python
# Install the Hugging Face Transformers library
!pip install transformers

# Import the pipeline function
from transformers import pipeline

# Create the pipeline
pipe = pipeline(model="facebook/bart-large-mnli")
```

```python
# Create a zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define the required labels for classification
req_labels = ['camera', 'battery', 'display', 'value for money', 'performance']

# Create an empty DataFrame to store the classification results
results_df = pd.DataFrame()

for index, row in df.iterrows():
    text_to_classify = row['Cleaned_Review']  # Replace 'text' with the actual column name in your DataFrame
    result = classifier(text_to_classify, req_labels)

    # Extract the labels and scores from the result
    labels = result['labels']
    scores = result['scores']

    # Create a dictionary to store the classification result for this text entry
    result_dict = {'Text': text_to_classify}

    # Populate the dictionary with the label scores
    for label, score in zip(labels, scores):
        result_dict[label] = score

    # Append the result to the results DataFrame
    results_df = pd.concat([results_df, pd.DataFrame(result_dict, index=[0])], ignore_index=True)

```

```python
results = []

for index, row in df.iterrows():
    text_to_classify = row['Cleaned_Review']  # Replace 'text' with the actual column name in your DataFrame
    result = classifier(text_to_classify, req_labels)

    # Extract the most likely label from the result
    most_likely_label = result['labels'][0]

    # Store the classification result in a dictionary
    result_dict = {
        'Text': text_to_classify,
        'Predicted Label': most_likely_label
    }

    # Append the result to the list
    results.append(result_dict)

# Create a new DataFrame to store the classification results
results_df = pd.DataFrame(results)

# Save the classification results to a CSV file
results_df.to_csv('classification_results.csv', index=False)
```
